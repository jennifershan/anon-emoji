```python
wc = WordCloud(font_path = './TwitterColorEmoji-SVGinOT.ttf', width = 1000, height = 800, prefer_horizontal = 1,
               regexp = emoji, collocations = False, colormap = 'BrBG').generate(aemojis)
wc.to_file('wc_anon_emojis' + '.png')

plt.figure(figsize = (20, 10))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```
